```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests,json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

data = json.loads(response)
print(data)
#json.loads()ç”¨æ–¼å°‡strå‹åˆ¥çš„è³‡æ–™è½‰æˆdictã€‚

print("ä¸€æ¬¡å‚³å›",len(data),"ç­†è³‡æ–™")
print("æ¯ä¸€ç­†è³‡æ–™åŒ…å«ä»¥ä¸‹æ¬„ä½",data[0].keys())
#å–dictçš„keyå€¼
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for i in range(len(data)):
    print("ç¬¬",i+1,"ç­†çš„æ¨™é¡Œ",(data[i]["title"]))
    print("ç¬¬",i+1,"ç­†çš„è²¼æ–‡æ™‚é–“",(data[i]["createdAt"]))
    print("ç¬¬",i+1,"ç­†çš„ç•™è¨€äººæ•¸",(data[i]["commentCount"]))
    print("ç¬¬",i+1,"ç­†çš„æŒ‰è®šäººæ•¸",(data[i]["likeCount"]))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

total_comment=0
total_like=0
count_number=0

for i in range(len(data)):
    total_comment=total_comment+data[i]["commentCount"]
    total_like=total_like+data[i]["likeCount"]
    count_number=count_number+1

print("ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸",total_comment/count_number)
print("ç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸",total_like/count_number)


import requests,json
f = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = f.text
#éç†±é–€æ–‡ç« è³‡æ–™

data_f = json.loads(response)
#json.loads()ç”¨æ–¼å°‡strå‹åˆ¥çš„è³‡æ–™è½‰æˆdictã€‚

total_comment=0
total_like=0
count_number=0

for i in range(len(data_f)):
    total_comment=total_comment+data_f[i]["commentCount"]
    total_like=total_like+data_f[i]["likeCount"]
    count_number=count_number+1

print("éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸",total_comment/count_number)
print("éç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸",total_like/count_number)
```
