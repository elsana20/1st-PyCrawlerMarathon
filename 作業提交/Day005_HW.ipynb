## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests,json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

data = json.loads(response)
print(data)
#json.loads()用於將str型別的資料轉成dict。

print("一次傳回",len(data),"筆資料")
print("每一筆資料包含以下欄位",data[0].keys())
#取dict的key值
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for i in range(len(data)):
    print("第",i+1,"筆的標題",(data[i]["title"]))
    print("第",i+1,"筆的貼文時間",(data[i]["createdAt"]))
    print("第",i+1,"筆的留言人數",(data[i]["commentCount"]))
    print("第",i+1,"筆的按讚人數",(data[i]["likeCount"]))
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

total_comment=0
total_like=0
count_number=0

for i in range(len(data)):
    total_comment=total_comment+data[i]["commentCount"]
    total_like=total_like+data[i]["likeCount"]
    count_number=count_number+1

print("熱門文章平均留言人數",total_comment/count_number)
print("熱門文章平均按讚人數",total_like/count_number)


import requests,json
f = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = f.text
#非熱門文章資料

data_f = json.loads(response)
#json.loads()用於將str型別的資料轉成dict。

total_comment=0
total_like=0
count_number=0

for i in range(len(data_f)):
    total_comment=total_comment+data_f[i]["commentCount"]
    total_like=total_like+data_f[i]["likeCount"]
    count_number=count_number+1

print("非熱門文章平均留言人數",total_comment/count_number)
print("非熱門文章平均按讚人數",total_like/count_number)
```
