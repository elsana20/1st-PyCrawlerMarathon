## requests + BeautifulSoup


````python
import requests
from bs4 import BeautifulSoup
url = 'http://www.qinxiaoshuo.com/read/0/2031/5dd1a01556fec86851279952.html'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text[0:3000])

````

````python
soup = BeautifulSoup(r.text, "html5lib")
print(soup)
````

## Grab + PyQuery

````python
from grab import Grab
g = Grab()
resp = g.go(url)
resp.body
````

````python
from pyquery import PyQuery
doc = PyQuery(resp.body)
title = doc('title')
print(type(title), title.text())
````
