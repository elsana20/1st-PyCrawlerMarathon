### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
#print(soup)

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find('em').getText(),d.find(class_="date").text, d.find_all('a')[0].text)
    #print(d.getText())
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news_data = {} #建立一個字典,用字典儲存

for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d.find('em').getText())

    tag = d.find('em').getText()
    #print(tag)
    #print(d.find_all('a')[0].text)
    if tag in news_data:
        news_data[tag].append(d.find(class_="date").text + " " + d.find_all('a')[0].text)
        #已經有tag值，新聞就用append的
    else:
        news_data[tag] = [d.find(class_="date").text + " " + d.find_all('a')[0].text]
        #沒有tag值，就建一個新聞資料
        

for tag in news_data:
    print(tag,"有",len(news_data[tag]),"筆資料")
    for d in news_data[tag]:
        print(d)
    print('\n')
```

### ③ 哪一個類別的文章最多

```python
news_max=0
max_tag=""

for tag in news_data:
    if len(news_data[tag])>news_max:
        news_max=len(news_data[tag])
        max_tag=tag          
print(max_tag,"類別的文章有",news_max,"篇，數量最多")
```
